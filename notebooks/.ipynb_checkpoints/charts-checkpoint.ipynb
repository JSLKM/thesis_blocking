```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=((rect.get_x() + rect.get_width() / 2), height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
def plot_chart(list_embedding, list_clustering, title):
    labels = ['RR', 'PC', 'alpha']
    x = np.arange(len(labels))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, list_embedding, width, label='embedding')
    rects2 = ax.bar(x + width/2, list_clustering, width, label='clustering')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Scores')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(loc=4)

    autolabel(rects1, ax)
    autolabel(rects2, ax)

    fig.tight_layout()

    plt.show()
```

```python
list_embedding = [0.9432, 0.9792, 0.9608]
list_clustering = [0.9790, 0.9464, 0.9624]
title = 'Scores Restaurant'

plot_chart(list_embedding, list_clustering, title)
```
