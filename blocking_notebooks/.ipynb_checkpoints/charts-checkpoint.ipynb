```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=((rect.get_x() + rect.get_width() / 2), height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
def plot_chart(list_RR, list_PC, list_alpha, title, saveName):
    labels = ['Standard', 'Suffix', 'Sorted Neighbourhood', 'Q-gram', 'Canopy', 'Emb-Cluster']
    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width*4/3, list_RR, width, label='RR - Reduction Ratio')
    rects2 = ax.bar(x, list_PC, width, label='PC - Pair Completeness')
    rects3 = ax.bar(x + width*4/3, list_alpha, width, label='Î± - Alpha')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Results')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend(loc=3)

    autolabel(rects1, ax)
    autolabel(rects2, ax)
    autolabel(rects3, ax)

    fig.tight_layout()
    fig.set_size_inches(18.5, 10.5)
    if saveName:
        fig.savefig(saveName, dpi=100)
    plt.show()
```

```python
list_RR = [0.9848,0.99,0.9848,0.9848,0.9848,0.9859]
list_PC = [0.9464,0.875,0.9464,0.9464,0.9464,0.9464]
list_alpha = [0.9652,0.9289,0.9652,0.9652,0.9652,0.9657]
title = 'Datasets Restaurant'

plot_chart(list_RR, list_PC, list_alpha, title, 'blocking_restaurant')
```
